```python
import pandas as pd
import numpy as np
```

```python
rng = np.random.default_rng()
```

```python
ids = np.arange(1, 11)
df = pd.DataFrame(
    data={
        "A": pd.Series(rng.random(10), dtype="float32", index=ids),
        "B": pd.Series(
            rng.choice(["strange", "charmed"], 10),
            dtype=pd.CategoricalDtype(categories=["strange", "charmed"]),
            index=ids,
        ),
    }
)
df
```

```python
for row in df.iterrows():
    print(type(row), len(row), type(row[1]))
    print(row[1].index)
    print(row[1])
```
