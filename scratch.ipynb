```python
from datetime import datetime, timedelta
import numpy as np
```

```python
rng = np.random.default_rng()
```

```python
start = datetime(year=2026, month=1, day=2)
```

```python
end = start + timedelta(days=53)
end
```

```python
timestamps = sorted(
    [
        int(
            (
                start + timedelta(days=day.item(), hours=hr.item(), minutes=min_.item())
            ).timestamp()
        )
        for day, hr, min_ in rng.integers(0, 54, (10, 3))
    ]
)
```

```python
for timestamp in timestamps:
    print(timestamp, datetime.fromtimestamp(timestamp))
```
