```python
from design.mockups.db.analysis import load_metrics
from datetime import datetime
import pandas as pd
import sqlite3 as sql
from contextlib import closing
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
DB_PATH = os.environ["DB_PATH"]
```

```python
with closing(sql.connect(DB_PATH)) as conn:
    mood = load_metrics(conn, 2, "Mood", "US/Pacific")
    quark = load_metrics(conn, 2, "Quark", "US/Pacific")
    meditate = load_metrics(conn, 2, "Meditate", "US/Pacific")
    water = load_metrics(conn, 2, "Water", "US/Pacific")
    weight = load_metrics(conn, 2, "Weight", "US/Pacific")
    exercise = load_metrics(conn, 2, "Exercise", "US/Pacific")
    
```

```python
today = pd.Timestamp.now(tz="US/Pacific").normalize()
today
```

```python
one_week_ago = today - pd.DateOffset(days=6)
one_week_ago
```

```python
all_days = pd.date_range(one_week_ago, today, freq="D")
all_days
```

```python
recent_weight = weight[weight["recorded_at"] >= one_week_ago]
recent_weight
```

```python
recent_weight["value"].mean()
```

```python
recent_weight.resample("D", on="recorded_at").mean().dropna().reindex(all_days, fill_value=0)
```
